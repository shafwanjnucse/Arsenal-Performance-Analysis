```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Arsenal_Data (1).xlsx")
```

```python
print(df.columns)
```

```python
#KPIs
df["Goal_difference"] = df['GS_ARS']-df['GC_ARS']
df["Points_per_game"] = df['Points'] / df['Match_Played']
df["Win Percentage"] = df["Wins"] / df['Match_Played']
df["Draw Percentage"] = df["TD_ARS"] / df['Match_Played']
df["Loss Percentage"]  = df["TL_ARS"] / df['Match_Played']
```

```python
df.to_excel('arsenal_processed1.xlsx', index=False)
```

```python
print(df[['Year', 'Points_per_game', 'Win Percentage', 'Draw Percentage', 'Loss Percentage']])
```
